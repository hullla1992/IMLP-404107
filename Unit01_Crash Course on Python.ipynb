### 01 - 奇偶數辨別（Odd or even）
#### 當使用者輸入一個的數字，它能夠辨別奇偶，並輸出檢驗結果給使用者。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一個數字 <br>
輸出(Output):奇數或偶數

* 進階: 可以限制輸入一個介於一定範圍（例如 1 到 1000）的數字

````python
p = int(input('請輸入一個數字 (介於0 - 1000): ')) # input 得出的型態type 會是str

if p > 1000 or p < 0:
    print('錯誤！你必須輸入0 - 1000 的數字')
else:
    if p % 2 == 0:
        print('噢！你輸入的%d 是一個偶數'%(p))
    else:
        print('噢！你輸入的%d 是一個奇數'%(p))
````

### 02 - 填字遊戲（Mad libs game）
#### 使用者輸入任何字，它可以是名詞、形容詞、動詞、代名詞等。程式得到文字後，可自行排列，組成成一個模板故事。

> 題目：輸入一數字判斷奇數或偶數<br>
輸入(Input): 一連串依據說明輸入對應文字 <br>
輸出(Output): 模板故事

* 進階: 可以讓程式重複執行

````python
import random

n = int(input('請問你要執行幾次: '))
for i in range(n):
    
    while True:
        weather = input('今天的天氣是: (請輸入數字，1.晴天/ 2.陰天/ 3.雨天)')
        if weather == '1':
            weather = '晴天'
            break
        elif weather == '2':
            weather = '陰天'
            break
        elif weather == '3':
            weather = '雨天'
            break
        else:
            print('你輸入錯誤的代碼，請輸入正確的數字')
            weather = input('今天的天氣是: (請輸入數字，1.晴天/ 2.陰天/ 3.雨天)')

    while True:
       transportation = input('搭什麼交通工具去上班: (請輸入數字，1.搭公車/ 2.開車/ 3.搭捷運)')
       if transportation == '1':
           transportation = '搭公車'
           break
       elif transportation == '2':
           transportation = '開車'
           break
       elif transportation == '3':
           transportation = '搭捷運'
           break
       else:
           print('你輸入錯誤的代碼，請輸入正確的數字')
           transportation = input('搭什麼交通工具去上班: (請輸入數字，1.搭公車/ 2.開車/ 3.搭捷運)')

    story_template_a = '今天的天氣是「%s」，我決定「%s」去上班'%(weather, transportation)
    story_template_b = '我查了今天的天氣預報後，今天的天氣「%s」，所以我選擇「%s」去上班'%(weather, transportation)
    story_template_list = [story_template_a, story_template_b]
    output = random.choice(story_template_list)
    print(output)
    print('--------------------------------------------')
````

### 03 -  猜數字（Guess the number）
#### 請使用者猜一個介於範圍（例如 1 到 100）之間的數字，
#### 若使用者猜錯，就詢問他們想繼續玩還是退出；若使用者猜對，就顯示祝賀訊息，
#### 並統計使用者的嘗試次數。如果使用者輸入的數字超出設定範圍，就顯示錯誤提示。

![](data\guess.png)


````python
import random

guess_count = 0
max_v = 100
min_v = 0
a = random.randint(1, 100)

while True:
    b =int(input('請你輸入一個數字: (提示: 數字介於%d - %d)'%(min_v, max_v)))
    
    if b < 100 and b > 0: #如果猜的數字沒有在範圍內的話，印出訊息並終止遊戲
        if guess_count <= 2: #設定猜題次數為3
            guess_count += 1
            if a == b:
                print('賓果！恭喜你猜對了！答案是「', a, '」')
                print('你總共猜了%d 次'%(guess_count))
                break
            elif a > b:
                min_v = b
                print('猜得太小了，在大一點 (提示: 數字介於%d - %d)'%(min_v, max_v))
                print('你目前已經猜了', str(guess_count), ' 次')
                continue_game = input('要退出遊戲嗎？輸入''esc''來退出，或按任意鍵繼續')
                if continue_game == 'esc':
                    print('遊戲已退出')
                    print('很遺憾遊戲失敗了，你總共猜了%d 次'%(guess_count))
                    break
            elif a < b:
                max_v = b
                print('猜得太大了，在小一點 (提示: 數字介於%d - %d)'%(min_v, max_v))
                print('你目前已經猜了' + str(guess_count) + ' 次')
                continue_game = input('要退出遊戲嗎？輸入''esc''來退出，或按任意鍵繼續')
                if continue_game == 'esc':
                    print('遊戲已退出')
                    print('很遺憾遊戲失敗了，你總共猜了%d 次'%(guess_count))
                    break
        else:
            print('你已經猜超過%d 次囉，你輸了！'%(guess_count)) 
            print('答案是', a)
            break         
    else:
        print('你輸入的數字不在範圍內喔！遊戲結束！')
        break
````

````python
email = input('What is your email address?: ').strip()
a = email.split('@')

# Slice out the user name
username = a[0]

# Slice the domain name
domain_name = a[1]

# Format message
output = "Your username is '{}' and your domain name is '{}'".format(username,domain_name)
print(output)
print(username)
print(domain_name)

domain = {'gmail.com':'Google', 'yahoo.com.tw':'Yahoo', 'ntu.edu.tw':'臺大', 'hotmail.com.tw':'Hotmail', 'hotmail.com':'Hotmail'}
if domain_name in domain:
    output = "這是註冊在 {} 之下的 Email 地址'".format(domain[domain_name])
else:
    output = "這是在 {} 之下自定義域".format(domain_name)
    print(output)
````

### 06 - 及格名單（Pass list）
#### 請使用集合功能來完成各科級名單的判斷
```
米花市帝丹小學一年级B班正舉辦期中考試
數學及格的有：柯南、灰原、步美、美環、光彦
英文及格的有：柯南、灰原、丸尾、野口、步美
以上已列出全班所有人
```

>請分別列出<br>
數學及格但英文不及格的同學名單<br>
數學不及格但英文及格的同學名單<br>
兩者皆及格名單<br>
* Hint:差集(減法)、交集


````python
m={'柯南','灰原','步美','美環','光彦'}
e={'柯南','灰原','丸尾','野口','步美'}

# 交集
a=m.intersection(e)
print(type(a)) # 資料型態會是set
a1=list(a) # 轉換成list 做排序
a1.sort()
print(a1)

# 差集，數學及格但是英文不及格
b=m.difference(a)
b1=list(b)
b1.sort()
print(b1)

# 差集，英文及格但是數學不及格
c=e.difference(a)
c1=list(c)
c1.sort()
print(c1)

# 班上的全部人
d=m.union(e)
d1=list(d)
d1.sort()
print(d1)
````

* 進階: 把file的學號抓出來，另存成一個新的csv檔

````python
import os

# 請使用者輸入或指定要查詢的路徑
yourPath = input('輸入: testfile')

# 列出指定路徑底下所有檔案(包含資料夾)
allFileList = os.listdir(yourPath) # list directory
print(allFileList)
print(type(allFileList)) # list 有順序性
ID = [] # 空的list 來存放學號

'''

輸入相對路徑: data
輸出: ['math_list.csv', 'poetry.txt', '05.JPG', 'guess.png', 'english_list.csv']

輸入絕對路徑: /Users/Unit01/data
輸出: ['math_list.csv', 'poetry.txt', '05.JPG', 'guess.png', 'english_list.csv']

'''

for file in allFileList: # 逐一查詢檔案清單
# 這邊也可以視情況，做檔案的操作(複製、讀取...等)
    if os.path.isdir(os.path.join(yourPath, file)): # 1. 使用isdir 檢查是否為目錄 | 2. 使用join 的方式把路徑與檔案名稱串起來(等同filePath+fileName)
        print("I'm a directory: " + file)
    # 使用isfile判斷是否為檔案
    elif os.path.isfile(os.path.join(yourPath, file)): # isfile 判斷是不是檔案
        print("I'm a File: " + file)
        # 學號的判斷
        print(file[file.index('_')+1:file.index('.')])

        # 判斷如果繳交的檔案格式不是txt，印出檔案名稱和錯誤訊息
        file_ext = file[file.index('.')+1:]
        if file_ext == 'txt':
            ID.append(file[file.index('_')+1:file.index('.')])
        else:
            print(file + ' 的檔案格式錯誤，檔案格式是' + file_ext)

print('檔案格式正確的學號，存在以下的list: ')
print(ID)
````

### 08 - 資料夾管理(shutil)
#### 說明 ####

```
在目前所在的目錄下建立一files資料夾

令使用者輸入一數字N，並在files資料夾中建立f1, f2… fN等N個資料夾後列出files的資料夾內容

將files資料夾裡的f1資料夾重新命名成folder1後再列出files的資料夾內容

移除files資料夾中的folder1後再列出files的資料夾內容

最後移除files資料夾
※須先退出files資料夾(os.chdir(../)) 
```


````python
import os
import shutil # shell utilities

# 檢查有沒有files 這個資料夾，如果有的話刪除，最後重新建立一個files 資料夾
if os.path.exists('files'):
    shutil.rmtree('files')
os.mkdir('files') # make directory


# 使用者輸入一數字n，並在files 資料夾中建立f1, f2... fn 等n 個資料夾後列出files 的資料夾內容
n = int(input('輸入要建立幾個資料夾: '))
os.chdir('files') #change directory

for i in range(1, n+1):
    os.mkdir('f' + str(i))
print(sorted(os.listdir()))
a = input('n 個資料夾建立了，按任意鍵繼續')

# 將files 資料夾裡的f1 資料夾重新命名成folder1 後再列出files 的資料夾內容
os.rename('f1', 'folder1')
print(sorted(os.listdir()))
a = input('f1 重新命名成folder1 了，按任意鍵繼續')

# 移除files 資料夾中的folder1 後再列出files 的資料夾內容
os.rmdir('folder1')
print(sorted(os.listdir()))
a = input('folder1 被移除了，按任意鍵繼續')

# 退出files 資料夾，然後移除files 資料夾
os.chdir('../') #../ 兩個點斜線是上一層，./ 一個點斜線是同一層
shutil.rmtree('files')


````

````python
def convertInputString():
    #請使用者輸入一串字串，先轉換字串為統一小寫
    rawInput = input("\nPlease enter a word, phrase, or a sentence \nto check if it is a palindrome: ")
    rawString = rawInput.lower() # 輸入轉為小寫
    rawList = list(rawString)
    return rawList

def stripAnalphabetics(dirtyList): 
    analphabeticList = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
    #去除相關標點符號
    for char in analphabeticList:
        if char in dirtyList:
            dirtyList.remove(char)
            return stripAnalphabetics(dirtyList) # 自己呼叫自己，直到所有標點符號都被移除進入下一行
    return dirtyList

def runPalindromeCheck(straightList):
    reversedList = straightList[::-1] # 顛倒過來
    if reversedList == straightList: 
        return "The text you have entered is a palindrome! %s"%(''.join(straightList)) 
    else: 
        return "The text you have entered is not a palindrome. %s"%(''.join(straightList)) 

def main(): 
    print("\nPalindrome checker")
    originalList = convertInputString()
    originalList = stripAnalphabetics(originalList)
    PalindromeCheck = runPalindromeCheck(originalList)
    print(PalindromeCheck)

main() # def 的函示不會被執行，直到被呼叫，呼叫的方式是函式名稱加小括號
````
